```python
import requests
from bs4 import BeautifulSoup
import json
import re
from time import sleep
from random import sample

HEADERS = {"User-Agent": "Mozilla/5.0"}

def get_athlete_data(url):
    try:
        response = requests.get(url, headers=HEADERS, timeout=15)
        soup = BeautifulSoup(response.text, "html.parser")


        # 👤 Prénom (.vn) et nom (.nn)
        prenom = nom = ""
        legend = soup.find("div", class_="legende")
        if legend:
            vn = legend.find("div", class_="vn")
            nn = legend.find("div", class_="nn")
            prenom = vn.text.strip() if vn else ""
            nom = nn.text.strip() if nn else ""

        # 🌍 Pays : img[title] dans .geburt
        country = None
        birth_div = soup.find("div", class_="geburt")
        if birth_div:
            flag = birth_div.find("img", title=True)
            if flag:
                country = flag["title"].strip()

# 🏅 Médailles
        medals = {"gold": 0, "silver": 0, "bronze": 0}
        total = 0
        teaser = soup.select_one("div.rnd.teaser")

        if teaser:
            blocks = teaser.find_all("div", recursive=False)  # enfants directs

            for block in blocks:
                mal = block.find("span", class_="mal")
                medal_div = block.find("div", class_="the-medal")

                if not mal:
                    continue

                match = re.search(r"\d+", mal.get_text(strip=True))
                count = int(match.group()) if match else 0

                if medal_div and medal_div.has_attr("data-medal"):
                    medal_type = medal_div["data-medal"]
                    if medal_type == "1":
                        medals["gold"] += count
                    elif medal_type == "2":
                        medals["silver"] += count
                    elif medal_type == "3":
                        medals["bronze"] += count

        total = sum(medals.values())

        return {
            "prenom": prenom,
            "nom": nom,
            "country": country,
            "medals": medals,
            "total": total,
            "url": url
        }

    except Exception as e:
        print(f"❌ Erreur pour {url}: {e}")
        return None


def scrape_all_athletes(urls, limit=1000):
    selected = sample(urls, min(limit, len(urls)))
    results = []

    for url in tqdm(selected, desc="Scraping athlètes", unit="it"):
        data = get_athlete_data(url)
        if data:
            results.append(data)
        sleep(0.3)

    print(f"✅ Scraping terminé : {len(results)} athlètes collectés.")
    return results
def export_to_json(data, filename="athletes.json"):
    with open(filename, "w", encoding="utf-8") as f:
        json.dump(data, f, indent=2, ensure_ascii=False)


all_athletes = scrape_all_athletes(liste_url_athletes, limit=1000)
export_to_json(all_athletes)
```

```python
import json
import pandas as pd
import matplotlib.pyplot as plt

# Charger le fichier JSON (dans le même dossier que ce script)
with open("countries.json", "r", encoding="utf-8") as f:
    data = json.load(f)

# Transformer les données : filtrer ceux avec un nom valide
records = [
    {"country": v["name"], "gold": v["gold"], "silver": v["silver"], "bronze": v["bronze"]}
    for v in data.values() if v["name"]
]
df = pd.DataFrame(records)

# Ajouter une colonne "total" (total médailles)
df["total"] = df["gold"] + df["silver"] + df["bronze"]

# Sélection des 100 pays avec le plus de médailles
top_100 = df.sort_values(by="total", ascending=False).head(100)

# Affichage graphique empilé
plt.figure(figsize=(20, 10))
plt.bar(top_100["country"], top_100["gold"], label="Gold")
plt.bar(top_100["country"], top_100["silver"], bottom=top_100["gold"], label="Silver")
plt.bar(top_100["country"], top_100["bronze"], bottom=top_100["gold"] + top_100["silver"], label="Bronze")
plt.xticks(rotation=90, fontsize=8)
plt.ylabel("Nombre de médailles")
plt.title("Top 100 pays par nombre total de médailles olympiques")
plt.legend()
plt.tight_layout()
plt.show()
```
